```python
from statsmodels.tsa.stattools import adfuller
print('Result of DickyFuller Test:')
def ad_test(weather_data):
    dftest = adfuller(weather_data, autolag='AIC')
    print("ADF: ", dftest[0])
    print("P Values: ", dftest[1])
    print("Lags: ", dftest[2])
    print("No. Of Ovservation: ", dftest[3])
    print("Critical Values: " )
    for key, val in dftest[4].items():
        print("\t",key, ": ", val )
```

```python
ad_test(weather_data['temperature'])
```

```python
from statsmodels.tsa.arima_model import ARIMA as ar
import warnings
```

```python
print(weather_data.shape)
train=weather_data.iloc[:-30]
test=weather_data.iloc[-30:]
print(train.shape,test.shape)
```

```python
model=ar(train['temperature'], order=(1,0,5))
model = model.fit()
model.summary()
```

```python
start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start, end=end)
pred.index=weather_data.index[start:end+1]
print(pred)
```

```python
pred.plot(legend=True)
test['temperature'].plot(legend=True)
```
